##### 1. Write a function which will take a string and return you the length of the string without using the inbuilt function len()

```python
def str_len(string):
    
    """This function gives you the length of a string"""
    
    if type(string) == str:
        length = 0
        
        for _ in string:
            length = length + 1
        return length
```

##### 3. Write a fucntion which will take the input as a dictionary and return all the values of the dict, even incase of a nested dictionary 

## Very IMPORTANT APPROACH

```python
def get_values(dictionary):
   
    """This function will take input as a dictionary and return all the vbalues of the dictionary, even in case of a nested dictionary"""

    values_list = []

    for value in dictionary.values():
        
        if isinstance(value, dict):
            values_list.extend(get_values(value))   # The value if dictionary, gets passed inside the function again to unwrap
                                                    # and gets gets passed into the list as parsed values from the resultant list.
        else:
            values_list.append(value)   # If not dict, the values gets appended on to the list
    
    return values_list
```

```python
x = {'a':"saheen", 'b':"ahzan", 'c':{"city":"TVM","state":{"code":"KL","name":"Kerala"},"country":{"code":"IN","name":"INDIA"}}}

get_values(x)
```

##### 5. Write a function which will take a list as an input and returns the concatination of all the elements as the output

```python
def concatinate_list(list1):

    """ This function which will take a list as an input and returns the concatination of all the elements as the output """

    concat = ""

    if isinstance(list1, list):
        for i in range(len(list1)):
            if isinstance(list1[i], int|float):
                concat = concat + str(list1[i])
            else:
                concat = concat + list1[i]
    return concat
```

##### 8. Write a function which will show the local system configuration.

```python
def show_system_configuration():
    
    """Prints the system configurations"""
    
    import platform
    import psutil           # A cross-platform module on system processes and system utilization

    print("Operating system:", platform.system())
    print("Release Version:", platform.release())
    print("System Version:", platform.version())
    print("Machine Type:", platform.machine())
    print("Processor:", platform.processor())
    
    ram = psutil.virtual_memory()
    print("Total RAM:", round(ram.total / (1024 ** 3),2), "GB")
```

```python
show_system_configuration()
```

##### 12. Write a function which can move a file from one directory to another directory.

```python
def move_file(source_file, dest_dir):

    """To move a file to another directory"""

    import shutil
    import os

    if os.path.isfile(source_file):
        shutil.move(source_file, os.path.join(dest_dir, source_file))
    print("Successfully moved the file")
```

##### 16. Write a function to read a pdf file.

```python
def read_pdf(file_path):
    # Open the PDF file in read binary mode
    
    import PyPDF2 as pdf

    with open(file_path, 'rb') as file:
        reader = pdf.PdfReader(file)

        for page in reader.pages:
            text = page.extract_text()
            print(text)
        
        file.close()
```

```python
read_pdf("comprehension.pdf")
```

#### 20. Write a function to request the user to enter an integer value. Unless the user enters an int value, request to try again.

```python
def get_int():
    
    flag = True
    while flag:
        
        try:
            integer = int(input("Enter an integer: "))
            if isinstance(integer, int):
                return "You have entered an integer", integer
                flag = False
        
        except Exception as err:
            print("No, that not an integer, try again. Error : ", err)
```

```python
get_int()
```
